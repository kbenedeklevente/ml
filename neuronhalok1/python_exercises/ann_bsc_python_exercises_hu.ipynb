## **0. feladat**

Definiáld a `relu` függvényt, ami a kapott számon a ReLU (Rectified Linear Unit) műveletet hajtja végre.

ReLU(z) = max(0,z)

Pl.:
```
5 -> 5
0.2 -> 0.2
-2. -> 0.
0. -> 0.
```


````
# place of solution

def relu(x):
  return 0 if x < 0 else x

# Tests
assert relu(5) == 5
assert relu(0.2) == 0.2
assert relu(-2.) == 0.
assert relu(0.) == 0.
print("Tests were successful.")
````
