ReLU(z) = max(0,z)

Pl.:
```
5 -> 5
0.2 -> 0.2
-2. -> 0.
```

Pl.:
```
mult_table(0) -> [[0]]
mult_table(1) -> [[0, 0], [0, 1]]
mult_table(2) -> [[0, 0, 0], [0, 1, 2], [0, 2, 4]]

mult_table(6)[4][5] -> 20
```

````
# place of solution

def mult_table(n):
  mtable = []
  for i in range(n+1):
    row = []
    mtable.append(row)
    for j in range(n+1):
      row.append(i*j)
  return mtable

# Tests
assert mult_table(0) == [[0]]
assert mult_table(1) == [[0, 0], [0, 1]]
assert mult_table(2) == [[0, 0, 0], [0, 1, 2], [0, 2, 4]]
assert mult_table(6)[4][5] == mult_table(6)[5][4] == 20
assert mult_table(10)[0][10] == 0
assert mult_table(10)[10][10] == 100
print("Tests were successful.")
````

Pl.:
```
fibonacci(0) -> 0
fibonacci(1) -> 1
fibonacci(n) -> fibonacci(n-1) + fibonacci(n-2)
```

````
# place of solution

# EFFICIENT SOLUTION: O(n) (linear) complexity
def fibonacci(n):
  series = [0,1]
  for i in range(2,n+1):
    series.append(series[-1] + series[-2])
  return series[n]

# Tests
assert fibonacci(0) == 0
assert fibonacci(1) == 1
for test_i in range(2,10):
  assert fibonacci(test_i) == fibonacci(test_i-1) + fibonacci(test_i-2)
print("Tests were successful.")
````

````
# place of solution

# INEFFICIENT SOLUTION: O(2^n) (exponential) instead of O(n) (linear) complexity
def fibonacci(n):
  return fibonacci(n-1) + fibonacci(n-2) if n >= 2 else n

# Tests
assert fibonacci(0) == 0
assert fibonacci(1) == 1
for test_i in range(2,10):
  assert fibonacci(test_i) == fibonacci(test_i-1) + fibonacci(test_i-2)
print("Tests were successful.")
````

````
# place of solution

# INEFFICIENT SOLUTION: O(n^2) instead of O(n) complexity
fibonacci = lambda n: fibonacci(n-1) + fibonacci(n-2) if n >= 2 else n

# Tests
assert fibonacci(0) == 0
assert fibonacci(1) == 1
for test_i in range(2,10):
  assert fibonacci(test_i) == fibonacci(test_i-1) + fibonacci(test_i-2)
print("Tests were successful.")
````

##  **6. feladat**

Definiáld a `is_paren_correct` függvényt, ami két-két féle nyitó- és csukózárójel (`()` és `[]`) esetén megadja, hogy helyes-e a zárójelek sorrendje a paraméterként kapott sztring kifejezésben.

Pl.:
```
is_paren_correct("") -> True
is_paren_correct("([]())[[([]())]]") -> True
is_paren_correct("([)]") -> False
is_paren_correct("(") -> False
```

````
# place of solution

def is_paren_correct(expr):
  pstack = []
  for ch in expr:
    if ch == ')' and (len(pstack) == 0 or pstack[-1] != '('):
      return False
    elif ch == ']' and (len(pstack) == 0 or pstack[-1] != '['):
      return False
    elif ch == ']' or ch == ')':
      pstack.pop()
    elif ch == '[' or ch == '(':
      pstack.append(ch)
  return len(pstack) == 0   # return True if empty

# Tests
assert is_paren_correct("")
assert is_paren_correct("()")
assert is_paren_correct("[]")
assert is_paren_correct("([]())[[([]())]]")
assert not is_paren_correct("([)]")
assert not is_paren_correct("(")
assert not is_paren_correct(")")
assert not is_paren_correct("[)")
assert not is_paren_correct("(]")
print("Tests were successful.")
````

## **7. feladat**

Definiáld a `triangle_area` függvényt, ami egy listányi számhármasról visszadja egyenként, hogy alkothatók-e háromszögek a számhármasok által megadott oldalhosszakból és ha igen, mennyi a háromszögek terüelte. A függvény visszatérési értéke egy lista, melynek i. eleme az input listában az i. számhármasból alkotott háromszög területe, illetve, ha nem alkotható háromszög, akkor pedig `None`.

A háromszög területének kiszámolására a Hérón-képlet alkalamzható. Egy számhármasból csak akkor alkotható háromszög, ha minden oldal hossza pozitív és teljesül a háromszög egyenlőtlenség.

Pl.:
```
triangle_area([]) -> []
triangle_area([(5,3,4), (4,1,2)]) -> [6, None]
```

````
import math

# place of solution

def triangle_area(triples):
  areas = []
  for a, b, c in triples:
    if (a >= b+c) or (b >= a+c) or (c >= a+b):
      areas.append(None)
    else:
      s = (a+b+c)/2.
      areas.append(math.sqrt(s*(s-a)*(s-b)*(s-c)))
  return areas

# Tests
assert triangle_area([]) == []
test_item1, test_item2 = triangle_area([(5,4,3), (4,1,2)])
assert (test_item1 - 6.) < 0.0001
assert test_item2 is None
print("Tests were successful.")
````

## **9. feladat**

Definiáld a `remaining_str` függvényt, ami egy sztringet és számpárok listáját kapja inputként. A visszatérési értéke a sztringnek az a része, amiből kivágtuk a számpárok által megjelölt intervallumokat. A számpárok első tagja a sztringből kivágandó rész első karakterének indexe, míg a számpár második tagja a kivágandó rész utáni első karakter indexe. Feltételezzük, hogy az összes számpár valóban létező indexekre mutat a sztringben.

Pl.:
```
remaining_str("abcdef", []) -> "abcdef"
remaining_str("abcdef", [(0,3)]) -> "def"
remaining_str("abcdef", [(0,3), (4,5)]) -> "df"
remaining_str("abcdef", [(2,5), (0,1)]) -> "bf"
```

````
# place of solution

def remaining_str(my_str, intervals_to_cut):
  remaining_idxs = [True]*len(my_str)
  for interval_start, interval_end in intervals_to_cut:
    for idx in range(interval_start, interval_end):
      remaining_idxs[idx] = False

  remaining_chars = []
  for idx in range(len(my_str)):
    if remaining_idxs[idx]:
      remaining_chars.append(my_str[idx])
  return ''.join(remaining_chars)


# Tests
assert remaining_str("abcdef", []) == "abcdef"
assert remaining_str("abcdef", [(0,3)]) == "def"
assert remaining_str("abcdef", [(0,3), (4,5)]) == "df"
assert remaining_str("abcdef", [(2,5), (0,1)]) == "bf"
print("Tests were successful.")
````

````
# place of solution

def remaining_str(my_str, intervals_to_cut):
  idxs_to_cut = set().union(*[set(range(int_start, int_end)) for int_start, int_end in intervals_to_cut])
  return ''.join([my_str[idx] for idx in range(len(my_str)) if idx not in idxs_to_cut])

  # [set(range(...))] list comprehension creates a list of sets of indices to be cut from the string
  # we take the union of these sets, with the set.union(<other_sets>) method
  # set.union() accepts any number of arguments (varargs)
  # we unpack the list of sets into many arguments for the set.union() method with the * unpack operator

# Tests
assert remaining_str("abcdef", []) == "abcdef"
assert remaining_str("abcdef", [(0,3)]) == "def"
assert remaining_str("abcdef", [(0,3), (4,5)]) == "df"
assert remaining_str("abcdef", [(2,5), (0,1)]) == "bf"
print("Tests were successful.")
````

Egy szótárban tárolják az egyes személyeknek megfelelő időpontokat név: számok listája formában. Definiáld az `acceptable_times` függvényt, ami az említett szótárak alapján visszatér időpontok listájával melyek mindenkinek megfelelnek. Ha nincs ilyen időpont, akkor üres listával tér vissza. Az időpontokat az egyszerűség kedvéért egy-egy szám jelöli.


Pl.:
```
acceptable_times({}) -> []
acceptable_times({'Anna': [5,7,3,2]}) -> [2,3,5,7]
acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,6], "Cecil":[7,5,3,6], "Dorina":[2,1,3,5,6,9]}) -> [3,6]
acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,9], "Cecil":[7,5,3,6], "Dorina":[2,1,5,6,9]}) -> []
```

````
# place of solution

def acceptable_times(time_dict):
  if len(time_dict) == 0:
    return []
  else:
    dict_timelists = list(time_dict.values())
    acceptable_set = set(dict_timelists[0])
    for time_list in dict_timelists[1:]:
      acceptable_set = acceptable_set.intersection(set(time_list))
    return list(acceptable_set)

# Tests
assert acceptable_times({}) == []
assert acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,6], "Cecil":[7,5,3,6], "Dorina":[2,1,3,5,6,9]}) == [3,6]
assert acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,9], "Cecil":[7,5,3,6], "Dorina":[2,1,5,6,9]}) == []
print("Tests were successful.")
````

````
# place of solution

def acceptable_times(time_dict):
  return [] if len(time_dict) == 0 else list(set(list(time_dict.values())[0]).intersection(*list(time_dict.values())))

# Tests
assert acceptable_times({}) == []
assert acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,6], "Cecil":[7,5,3,6], "Dorina":[2,1,3,5,6,9]}) == [3,6]
assert acceptable_times({'Anett':[3,5,6], "Barna":[1,3,2,9], "Cecil":[7,5,3,6], "Dorina":[2,1,5,6,9]}) == []
print("Tests were successful.")
````

Pl.:
```
same_magnitude([2,5,3,2]) -> False
same_magnitude([4,5,3,4]) -> True
same_magnitude([4,5,3,-2]) -> False
same_magnitude([0,0,0]) -> False
same_magnitude([-2,-2,-2]) -> False
```

````
# place of solution

def same_magnitude(my_list):
  limit = max(my_list)*.5
  for item in my_list:
    if (item <= 0) or (item <= limit):
      return False
  return True

# Tests
assert not same_magnitude([2,5,3,2])
assert same_magnitude([4,5,3,4])
assert not same_magnitude([4,5,3,-2])
assert not same_magnitude([0,0,0])
assert not same_magnitude([-2,-2,-2])
print("Tests were successful.")
````

````
# place of solution

def same_magnitude(my_list):
  limit = max(my_list)*.5
  return all([(item > limit) and (item > 0) for item in my_list])

# Tests
assert not same_magnitude([2,5,3,2])
assert same_magnitude([4,5,3,4])
assert not same_magnitude([4,5,3,-2])
assert not same_magnitude([0,0,0])
assert not same_magnitude([-2,-2,-2])
print("Tests were successful.")
````

````
# place of solution

def same_magnitude(my_list):
  limit = max(max(my_list)*.5, 0.)
  return not any(map(lambda item: item <= limit, my_list))

# Tests
assert not same_magnitude([2,5,3,2])
assert same_magnitude([4,5,3,4])
assert not same_magnitude([4,5,3,-2])
assert not same_magnitude([0,0,0])
assert not same_magnitude([-2,-2,-2])
print("Tests were successful.")
````

Ha a lista nem tartalmaz két elemet, akkor térjen vissza a függvény None-al, egyébként pedig a legutolsó végeredménnyel. (Ez a függvény a funkcionális nyelvekből ismert left fold).


Pl.:

```
my_reduce(+, [1]) -> None
```
